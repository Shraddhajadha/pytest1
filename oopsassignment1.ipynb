```python
#"Problem 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.\n",

class bank_account:

    def __init__(self,account_number,account_holder_name,balance):
        try:
            self.account_number=account_number
            self.account_holder_name=account_holder_name
            self.balance=balance
        except Exception as e:
            print(e)

    def deposite(self,deposite_amount):
        try:
            self.balance+=deposite_amount
            print("Your money is deposited successfully.")
            return self.balance
        except Exception as e:
            print(e)

    def withdraw(self,withdraw_amount):
        try:
            if withdraw_amount <=self.balance:
                self.balance-=withdraw_amount
                print("Your money is withdraw successfully.")
    
            else:
                print("Insufficient balance.")
            return self.balance
        except Exception as e:
            print(e)
```

```python
shraddha=bank_account(123,"shraddha",5000)
```

```python
shraddha.deposite(2000)
```

```python
shraddha.withdraw(1000)
```

```python
#"Problem 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.\n",
class vehicle_rental_system:

    def __init__(self,total_vehicle):
        self.total_vehicle=total_vehicle

    def display_available(self):
        if self.total_vehicle:
            print("available vehicles.")
            for v in self.total_vehicle:
                print("-",v)
        else:
            print("No vehicles available right now.")

    def rent_vehicle(self,vehicle_name):
        if vehicle_name in self.total_vehicle:
            self.total_vehicle.remove(vehicle_name)
            print(f"{vehicle_name} has been rented successfully.")

        else:
            print(f"Sorry, {vehicle_name} is not available.")

    def return_vehicle(self,vehicle_name):
        self.total_vehicle.append(vehicle_name)
        print(f"{vehicle_name} has been return successfully.")
```

```python
vehicle_system=vehicle_rental_system(["car","van","bus","scooter","bike"])
```

```python
vehicle_system.display_available()
```

```python
vehicle_system.rent_vehicle("car")
```

```python
vehicle_system.return_vehicle("car")
```

```python
#"Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.\n",
class product:
    def __init__(self,product_id,name,price,quantity):
        self.product_id=product_id
        self.name=name
        self.price=price
        self.quantity=quantity

    def __str__(self):
        return f"{self.product_id}: {self.name} - ${self.price}, Qty: {self.quantity}"

class inventory:
    def __init__(self,products=None):
        if products is None:
            products={}
        self.products={}

    def add_products(self,product):
        if product.product_id in self.products:
            print(f"Product {product.name} already exists. Updating quantity instead.")
            self.products[product.product_id].quantity+=product.quantity
        else:
            self.products[product.product_id]=product
            print(f"Product {product.name} added successfully.")

    def update_product_quantity(self,product_id,quantity):
        if product_id in self.products:
            self.products[product_id].quantity = quantity
            print(f"Quantity of {self.products[product_id].name} updated to {quantity}.")
        else:
            print("Product not found in inventory.")

    def display_available_products(self):
        """Display all available products"""
        if self.products:
            print("Available products:")
            for product in self.products.values():
                print("-", product)
        else:
            print("No products in inventory.")
```

```python
prod=product(5,"cream",50,100)
```

```python
invent=inventory()
```

```python
invent.add_products(prod)
```

```python
invent.update_product_quantity(5,105)
```

```python
invent.display_available_products()
```

```python
# "Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.\n",
class shape_calculation:
    def __init__(self,length,width,height=None):
        self.shape_length=length
        self.shape_width=width
        self.shape_height=height

    def area_of_shape(self):
        if self.shape_height is None:
            return self.shape_length*self.shape_width
        else:
            return 2*(self.shape_length*self.shape_width+self.shape_width*self.shape_height+self.shape_height*self.shape_length)

    def perimeter(self):
        if self.shape_height is None:
            return 2*(self.shape_length+self.shape_width)
        else:
            return self.shape_length*self.shape_width*self.shape_height
```

```python
shape=shape_calculation(5,3)
```

```python
shape.area_of_shape()
```

```python
shape.perimeter()
```

```python
shape_cuboid=shape_calculation(5,3,2)
```

```python
shape_cuboid.area_of_shape()
```

```python
shape_cuboid.perimeter()
```

```python
#"Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.\n",
class student:
    def __init__(self,student_id,name,grades):
        self.student_id=student_id
        self.name=name
        self.grades=grades

    def calculate_average(self):
        if self.grades:
            return sum(self.grades)/len(self.grades)
        else:
            return 0

    def display_student_details(self):
        print(f"Student ID : {self.student_id}")
        print(f"Name : {self.name}")
        print(f"Grades : {self.grades}")
        print(f"Student Average : {self.calculate_average():.2f}")
```

```python
stud.calculate_average()
```

```python
stud.display_student_details()
```

```python
#"Problem 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.\n",
class social_media:
    def __init__(self,username):
        self.username=username
        self.posts=[]

    def add_post(self,post):
        self.posts.append(post)
        print(f"{post} added successfully.")

    def display_post(self):
        if self.posts:
            print(f"\n {self.username}'s posts")
            for i,post in enumerate(self.posts,start=1):
                print(f"{i}.{post}")
        else:
            print(f"{self.username}'s has no posts yet.")

    def search_post(self,keyword):
        results=[post for post in self.posts if keyword.lower() in post.lower()]
        if results:
            print(f"\nPosts containing '{keyword}':")
            for post in results:
                print("-", post)
        else:
            print(f"No posts found with keyword '{keyword}'.")
```

```python
#"Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks."
class todo_list:
    def __init__(self):
        self.tasks=[]

    def add_tasks(self,task, due_date):
        self.tasks.append({"task":task,"due_date":due_date,"completed":False})
        print(f"Task '{task}' added with due date {due_date}")

    def mark_task(self,task):
        for t in self.tasks:
            if t["task"].lower()==task.lower():
                t["completed"]=True
                print(f"Task '{task}' marked as completed.")
                return
        print(f"Task '{task}' not found.")

    def display_pending_task(self):
        pending=[t for t in self.tasks if not t["completed"]]
        if pending:
            print("\n Pending tasks.")
            for p in pending:
                print(f"- {p['task']} (Due: {p["due_date"]})")
        else:
            print("No pending tasks. All caught up!")
```

```python
to_do=todo_list()
```

```python
to_do.add_tasks("Join the class at 9 AM","2025-07-05")
```

```python
to_do.add_tasks("solve the assignment today","2025-07-06")
```

```python
to_do.mark_task("Join the class at 9 AM")
```

```python
to_do.display_pending_task()
```
