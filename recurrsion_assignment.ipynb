```python
''' Q.1 Can you explain the logic and working of the Tower of Hanoi algorithm by writing a Java program?
How does the recursion work, and how are the movements of disks between rods accomplished? '''

def tower_hanoi(n,source,auxilary,destination):
    if n==1:
        print(f"Move 1st disk {source} to {destination}")
        return

    tower_hanoi(n-1,source,destination,auxilary)
    print(f"Move {n}th disk {source} to {destination}")
    tower_hanoi(n-1,auxilary,source,destination)
```

```python
tower_hanoi(3,'a','b','c')
```

```python
'''  Q.2 Given two strings word1 and word2, return the minimum number of operations required to convert word1
to word2. '''

def edit_distance(word1,word2,i,j):
    if i<0:
        return j+1

    if j<0:
        return i+1

    if word1[i]==word2[j]:
        return edit_distance(word1,word2,i-1,j-1)

    return 1+min(
        edit_distance(word1,word2,i,j-1),
        edit_distance(word1,word2,i-1,j),
        edit_distance(word1,word2,i-1,j-1)
    )
```

```python
#Q. 3 Print the max value of the array [ 13, 1, -3, 22, 5].
arr = [13, 1, -3, 22, 5]
def find_max(arr,n):
    if n==1:
        return arr[0]

    return max(arr[n-1],find_max(arr,n-1))
```

```python
find_max(arr,5)
```

```python
#Q.4 Find the sum of the values of the array [92, 23, 15, -20, 10].
arr=[92, 23, 15, -20, 10]
def sum_values(arr,n):
    if n==1:
        return arr[0]

    return arr[n-1]+sum_values(arr,n-1)
```

```python
sum_values(arr,5)
```

```python
''' Q.5 Given a number n. Print if it is an armstrong number or not.An armstrong number is a number if the sum
of every digit in that number raised to the power of total digits in that number is equal to the number.
Example : 153 = 1^3 + 5^3 + 3^3 = 1 + 125 + 27 = 153 hence 153 is an armstrong number. (Easy)
Input1 : 153
Output1 : Yes
Input 2 : 134
Output2 : No '''

def count_digit(n):
    if n==0:
        return 0
    return 1+count_digit(n//10)

def armstrong_sum(n,power):
    if n==0:
        return 0
    digit=n%10
    return (digit**power)+armstrong_sum(n//10,power)

def is_armstrong(n):
    digits=count_digit(n)
    return n==armstrong_sum(n,digits)

# Test cases
print("Yes" if is_armstrong(153) else "No")
print("Yes" if is_armstrong(134) else "No")
```
