```python
'''12. Explain the use of abstract base classes (ABCs) in Python and how they relate to inheritance. Provide an
example using the `abc` module.'''

#abstract base class means its not instantiated directly.There subclasses must be implement.
import math
from abc import ABC,abstractmethod
class shape(ABC):
    @abstractmethod
    def area(self):
        print("These method overidden in child classes.")


class circle(shape):
    def __init__(self,radius):
        self.radius=radius
        
    def area(self):
        return math.pi*self.radius**2

class rectangle(shape):
    def __init__(self,length,width):
        self.length=length
        self.width=width

    def area(self):
        return self.length*self.width

c=circle(5)
r=rectangle(4,6)

c.area()
c.area()
r.area()
```

```python
#18. What is the "diamond problem" in multiple inheritance, and how does Python address it?
#The diamond problem occurs when a class inherits from two classes that have a common ancestor, forming a diamond-shaped inheritance diagram.
#The problem arises when:

'''The child class calls a method that exists in the topmost ancestor.

Python (or other languages like C++) needs to decide which ancestor’s method to call.'''

class A:
    def show(self):
        print("Method in A")

class B(A):
    def show(self):
        print("Method in B")

class C(A):
    def show(self):
        print("Method in C")

class D(B, C):
    pass

d = D()
d.show()

'''Python uses C3 Linearization (Method Resolution Order) to determine the order in which base classes are searched.

You can check the MRO using:'''

print(D.mro())
```

```python
'''20. Create a Python class hierarchy for a university system. Start with a base class `Person` and create child
classes `Student` and `Professor`, each with their own attributes and methods. Provide an example of using
these classes in a university context.'''

class person:
    def __init__(self,name,age):
        self.name=name
        self.age=age

    def show_details(self):
        print(f"Name: {self.name}")
        print(f"Age: {self.age}")

class student(person):
    def __init__(self,name,age,student_id,major):
        super().__init__(name,age)
        self.student_id=student_id
        self.major=major

    def study(self):
        print(f"{self.name} is studying in {self.major}")

    def show_details(self):
        super().show_details()
        print(f"Student ID: {self.student_id}")
        print(f"Major: {self.major}")

class Professor(person):
    def __init__(self,name,age,department,salary):
        super().__init__(name,age)
        self.department=department
        self.salary=salary

    def teach(self):
        print(f"{self.name} is teaching in the {self.department} department.")

    def show_details(self):
        super().show_details()
        print(f"Department: {self.department}")
        print(f"Salary: ₹{self.salary}")

s=student("kau",31,25,'dsa')
p=Professor("mau",25,"computer",20000)
s.study()
s.show_details()
p.teach()
p.show_details()
```

```python
#12. Describe the use of abstract properties in Python and how they can be employed in abstract classes.

''' Abstract Properties in Python
What Are Abstract Properties?

Abstract properties are attributes in abstract classes that must be defined in subclasses.

They are a combination of properties and abstract methods.

Abstract properties allow you to define getters, setters, and deleters that must be implemented in any subclass.'''

from abc import ABC, abstractmethod

# Abstract class with abstract property
class Vehicle(ABC):

    @property
    @abstractmethod
    def speed(self):
        pass

    @abstractmethod
    def start(self):
        pass

# Subclass implementing the abstract property
class Car(Vehicle):
    def __init__(self, speed):
        self._speed = speed  # internal storage

    @property
    def speed(self):
        return self._speed

    @speed.setter
    def speed(self, value):
        if value > 0:
            self._speed = value
        else:
            print("Speed must be positive")

    def start(self):
        print(f"Car started at {self.speed} km/h")

# Using the class
car = Car(80)
car.start()
car.speed = 100
car.start()
```

```python
#13. Create a Python class hierarchy for employees in a company (e.g., manager, developer, designer) and implement abstraction by defining a common `get_salary()` method.
from abc import ABC, abstractmethod

# Abstract Base Class
class Employee(ABC):

    def __init__(self, name):
        self.name = name

    @abstractmethod
    def get_salary(self):
        pass  # Abstract method to be implemented by subclasses


# Subclass 1: Manager
class Manager(Employee):
    def __init__(self, name, base_salary, bonus):
        super().__init__(name)
        self.base_salary = base_salary
        self.bonus = bonus

    def get_salary(self):
        return self.base_salary + self.bonus


# Subclass 2: Developer
class Developer(Employee):
    def __init__(self, name, base_salary, overtime):
        super().__init__(name)
        self.base_salary = base_salary
        self.overtime = overtime

    def get_salary(self):
        return self.base_salary + self.overtime


# Subclass 3: Designer
class Designer(Employee):
    def __init__(self, name, base_salary, commission):
        super().__init__(name)
        self.base_salary = base_salary
        self.commission = commission

    def get_salary(self):
        return self.base_salary + self.commission


# Example Usage
employees = [
    Manager("Alice", 80000, 15000),
    Developer("Bob", 60000, 10000),
    Designer("Charlie", 50000, 5000)
]

for emp in employees:
    print(f"{emp.name}'s salary: {emp.get_salary()}")
```

```python
#15. Explain the concept of abstract data types (ADTs) and their role in achieving abstraction in Python.

''' What Is an Abstract Data Type (ADT)?

An Abstract Data Type (ADT) is a conceptual model of a data structure that defines the data and the operations allowed on it, without specifying how the operations are implemented internally.

In simple terms:

An ADT focuses on what operations can be performed, not how they are implemented.

Examples of ADTs include:

Stack

Queue

List

Dictionary

Set

These describe behavior, not implementation. '''

class Stack:
    def __init__(self):
        self.items = []

    def push(self, item):      # Operation
        self.items.append(item)

    def pop(self):             # Operation
        if not self.is_empty():
            return self.items.pop()
        return "Stack is empty"

    def peek(self):            # Operation
        if not self.is_empty():
            return self.items[-1]
        return None

    def is_empty(self):        # Operation
        return len(self.items) == 0
```

```python
#19. Create a Python class for a library system, implementing abstraction by defining common methods (e.g., `add_book()`, `borrow_book()`) in an abstract base class.
from abc import ABC, abstractmethod

# Abstract Base Class
class LibrarySystem(ABC):

    @abstractmethod
    def add_book(self, title, author):
        pass

    @abstractmethod
    def borrow_book(self, title):
        pass

    @abstractmethod
    def return_book(self, title):
        pass


# Concrete Subclass
class PublicLibrary(LibrarySystem):
    
    def __init__(self):
        self.books = {}          # {title: {"author": ..., "available": True/False}}

    def add_book(self, title, author):
        self.books[title] = {"author": author, "available": True}
        return f"Book '{title}' added to the library."

    def borrow_book(self, title):
        if title in self.books and self.books[title]["available"]:
            self.books[title]["available"] = False
            return f"You borrowed '{title}'."
        return f"'{title}' is not available."

    def return_book(self, title):
        if title in self.books and not self.books[title]["available"]:
            self.books[title]["available"] = True
            return f"Book '{title}' returned successfully."
        return f"'{title}' was not borrowed or does not exist."
        

# Example Usage
if __name__ == "__main__":
    library = PublicLibrary()

    print(library.add_book("1984", "George Orwell"))
    print(library.borrow_book("1984"))
    print(library.return_book("1984"))
```

```python
'''8. Create a Python class for a computer system, using composition to represent components like CPU, RAM,
and storage devices.'''

class CPU:
    def __init__(self,brand,cores,frequency_ghz):
        self.brand = brand
        self.cores = cores
        self.frequency_ghz = frequency_ghz

    def specs(self):
        return f"{self.brand} CPU, {self.cores} cores @ {self.frequency_ghz}GHz"

class RAM:
    def __init__(self, size_gb, ram_type):
        self.size_gb = size_gb
        self.ram_type = ram_type

    def specs(self):
        return f"{self.size_gb}GB {self.ram_type} RAM"

class Storage:
    def __init__(self, size_gb, storage_type):
        self.size_gb = size_gb
        self.storage_type = storage_type

    def specs(self):
        return f"{self.size_gb}GB {self.storage_type} storage"

class Computer:
    def __init__(self, cpu, ram, storage_devices):
        self.cpu = cpu
        self.ram = ram
        self.storage_devices = storage_devices 

    def system_specs(self):
        print("Computer Specifications:")
        print(self.cpu.specs())
        print(self.ram.specs())
        print("Storage Devices:")
        for storage in self.storage_devices:
            print("-", storage.specs())

cpu = CPU("Intel", 8, 3.6)
ram = RAM(16, "DDR4")
storage1 = Storage(512, "SSD")
storage2 = Storage(1000, "HDD")

computer = Computer(cpu, ram, [storage1, storage2])
computer.system_specs()
```

```python
'''10. Create a Python class for a car, using composition to represent components like the engine, wheels, and
transmission.'''

class Engine:
    def __init__(self, horsepower, fuel_type):
        self.horsepower = horsepower
        self.fuel_type = fuel_type

    def start(self):
        return "Engine started"

class Transmission:
    def __init__(self, transmission_type):
        self.transmission_type = transmission_type  # Manual / Automatic

    def shift_gear(self, gear):
        return f"Shifted to gear {gear}"

class Wheel:
    def __init__(self, size):
        self.size = size

    def rotate(self):
        return "Wheel is rotating"

class Car:
    def __init__(self, brand, engine, transmission, wheels):
        self.brand = brand
        self.engine = engine              # Composition
        self.transmission = transmission  # Composition
        self.wheels = wheels              # List of Wheel objects

    def drive(self):
        actions = [
            self.engine.start(),
            self.transmission.shift_gear(1),
        ]
        actions.extend(wheel.rotate() for wheel in self.wheels)
        return actions

# Example usage
engine = Engine(150, "Petrol")
transmission = Transmission("Automatic")
wheels = [Wheel(16) for _ in range(4)]

car = Car("Toyota", engine, transmission, wheels)

for action in car.drive():
    print(action)
```

```python
'''12. Create a Python class for a university course, using composition to represent students, instructors, and
course materials.'''

class Student:
    def __init__(self, student_id, name):
        self.student_id = student_id
        self.name = name

    def __str__(self):
        return f"{self.student_id} - {self.name}"

class Instructor:
    def __init__(self, name, department):
        self.name = name
        self.department = department

    def __str__(self):
        return f"{self.name} ({self.department})"


class CourseMaterial:
    def __init__(self, title, material_type):
        self.title = title
        self.material_type = material_type  # Book, Slides, Video

    def __str__(self):
        return f"{self.material_type}: {self.title}"

class course:
    def __init__(self,course_code,course_name):
        self.course_code=course_code
        self.course_name=course_name
        self.students=[]
        self.instructors=[]
        self.materials =[]

    def add_student(self, student):
        self.students.append(student)

    def add_instructor(self, instructor):
        self.instructors.append(instructor)

    def add_material(self, material):
        self.materials.append(material)

    def course_details(self):
        return {
            "Course": f"{self.course_code} - {self.course_name}",
            "Students": [str(s) for s in self.students],
            "Instructors": [str(i) for i in self.instructors],
            "Materials": [str(m) for m in self.materials],
        }

# Create students
s1 = Student(101, "Alice")
s2 = Student(102, "Bob")

# Create instructor
i1 = Instructor("Dr. Smith", "Computer Science")

# Create materials
m1 = CourseMaterial("Python Programming", "Book")
m2 = CourseMaterial("OOP Concepts", "Slides")

# Create course
course = course("CS101", "Introduction to Programming")

course.add_student(s1)
course.add_student(s2)
course.add_instructor(i1)
course.add_material(m1)
course.add_material(m2)

# Display course details
details = course.course_details()
for key, value in details.items():
    print(key, ":", value)
```

```python
'''14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes,
and ingredients.'''

class Ingredient:
    def __init__(self, name, quantity, unit):
        self.name = name
        self.quantity = quantity
        self.unit = unit

    def __str__(self):
        return f"{self.name} ({self.quantity} {self.unit})"


class Dish:
    def __init__(self, name, price):
        self.name = name
        self.price = price
        self.ingredients = []   # Composition: Dish has Ingredients

    def add_ingredient(self, ingredient):
        self.ingredients.append(ingredient)

    def __str__(self):
        return f"{self.name} - ${self.price:.2f}"


class Menu:
    def __init__(self, menu_name):
        self.menu_name = menu_name
        self.dishes = []        # Composition: Menu has Dishes

    def add_dish(self, dish):
        self.dishes.append(dish)

    def display_menu(self):
        print(f"\n{self.menu_name}")
        for dish in self.dishes:
            print(dish)


class Restaurant:
    def __init__(self, name):
        self.name = name
        self.menus = []         # Composition: Restaurant has Menus

    def add_menu(self, menu):
        self.menus.append(menu)

    def show_restaurant_menu(self):
        print(f"Welcome to {self.name}")
        for menu in self.menus:
            menu.display_menu()


# Ingredients
cheese = Ingredient("Cheese", 100, "grams")
tomato = Ingredient("Tomato", 2, "pieces")
dough = Ingredient("Dough", 1, "base")

# Dish
pizza = Dish("Margherita Pizza", 8.99)
pizza.add_ingredient(cheese)
pizza.add_ingredient(tomato)
pizza.add_ingredient(dough)

# Menu
main_menu = Menu("Main Menu")
main_menu.add_dish(pizza)

# Restaurant
restaurant = Restaurant("Good Food Place")
restaurant.add_menu(main_menu)

restaurant.show_restaurant_menu()
```

```python
'''20. Create a Python class for a social media application, using composition to represent users, posts, and
comments.'''

class Comment:
    def __init__(self, author, content):
        self.author = author
        self.content = content

    def __str__(self):
        return f"{self.author}: {self.content}"


class Post:
    def __init__(self, author, content):
        self.author = author
        self.content = content
        self.comments = []     # Composition: Post has Comments

    def add_comment(self, comment):
        self.comments.append(comment)

    def __str__(self):
        return f"Post by {self.author}: {self.content}"


class User:
    def __init__(self, username):
        self.username = username
        self.posts = []        # Composition: User has Posts

    def create_post(self, content):
        post = Post(self.username, content)
        self.posts.append(post)
        return post

    def __str__(self):
        return self.username


class SocialMediaApp:
    def __init__(self, name):
        self.name = name
        self.users = []        # Composition: App has Users

    def add_user(self, user):
        self.users.append(user)

    def show_all_posts(self):
        for user in self.users:
            for post in user.posts:
                print(post)
                for comment in post.comments:
                    print("  -", comment)

# Create users
alice = User("Alice")
bob = User("Bob")

# Create app
app = SocialMediaApp("MySocialApp")

app.add_user(alice)
app.add_user(bob)

# Users create posts
post1 = alice.create_post("Hello, world!")
post2 = bob.create_post("Learning Python!")

# Add comments
post1.add_comment(Comment("Bob", "Nice post!"))
post2.add_comment(Comment("Alice", "Keep going!"))

# Display posts and comments
app.show_all_posts()
```
