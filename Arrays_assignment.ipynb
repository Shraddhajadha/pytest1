```python
'''1. Given an array, check if it contains any duplicates or not.
arr = [1, 2, 4, 2, 5, 9]
Output = True'''
arr = [1, 2, 4, 2, 5, 9]
def check_duplicates():
    convert_list=set()
    for i in arr:
        if i in convert_list:
            return True
        convert_list.add(i)
    return False

check_duplicates()

#Time complexity=O(n)
#Space complexity=O(1)
```

```python
'''2. Given an array and an integer k, rotate the array to the right by k steps.
arr = [1, 2, 3, 4, 5, 6, 7] k = 3
Output = [5, 6, 7, 1, 2, 3, 4]'''

arr=[1, 2, 3, 4, 5, 6, 7]
k=3
def rotate_right(arr,k):
    n=len(arr)
    k=k%n
    arr.reverse()
    arr[:k]=reversed(arr[:k])
    arr[k:]=reversed(arr[k:])

    return arr

rotate_right(arr,k)

#Time complexity=O(n)
#Space complexity=O(1)
```

```python
'''3. Reverse the given array in-place, means without using any extra data structure.
arr = [2, 4, 5, 7, 9, 12]
Output = [12, 9, 7, 5, 4, 2]'''
arr=[2, 4, 5, 7, 9, 12]

def reverse_array(arr):
    left=0
    right=len(arr)-1

    while left<right:
        arr[left],arr[right]=arr[right],arr[left]
        left+=1
        right-=1

    return arr
reverse_array(arr)

#Time complexity=O(n)
#Space complexity=O(1)
```

```python
'''4. Given an array of integers, find the maximum element in an array
arr = [10, 5, 20, 8, 15]
Output = 20'''

arr=[10, 5, 20, 8, 15]

def max_element(arr):
    element=arr[0]
    for i in arr:
        if i > element:
            element=i

    return element

max_element(arr)

#Time complexity=O(n)
#Space complexity=O(1)
```

```python
'''5. Given a sorted array, remove the duplicate element without using any extra data structure.
arr = [1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 5, 5]
Output = [1, 2, 3, 4, 5]'''

arr=[1, 1, 2, 2, 2, 3, 3, 4, 4, 4, 5, 5]

def remove_duplicate(arr):
    if not arr:
        return arr
    j=0
    for i in range(len(arr)):
        if arr[i]!=arr[j]:
            j+=1
        arr[j]=arr[i]

    return arr[:j+1]

remove_duplicate(arr)

#Time complexity=O(n)
#Space complexity=O(1)
```
